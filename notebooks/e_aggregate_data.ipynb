```r
library(dplyr)
```

```r
dta <- read.csv('../data/seattle_911_geo_time_features.csv')
head(dta)
```

```r
cols <- c(seq(9, 22), 24)
dta <- subset(dta, select = cols)
head(dta)
```

```r
colnames(dta)
```

```r
dta1 <- subset(dta[dta$Zone == 1,], select = seq(1,14))
dta2 <- subset(dta[dta$Zone == 2,], select = seq(1,14))
dta3 <- subset(dta[dta$Zone == 3,], select = seq(1,14))
dta4 <- subset(dta[dta$Zone == 4,], select = seq(1,14))
dta5 <- subset(dta[dta$Zone == 5,], select = seq(1,14))
dta6 <- subset(dta[dta$Zone == 6,], select = seq(1,14))
dta7 <- subset(dta[dta$Zone == 7,], select = seq(1,14))
```

```r
head(dta1)
```

```r
counts1 <- dta1 %>% group_by_(.dots=colnames(dta1)) %>% summarise(Freq = n())
plot(x = counts1$DayNumber, y = counts1$Freq)
```

```r
counts2 <- dta2 %>% group_by_(.dots=colnames(dta2)) %>% summarise(Freq = n())
plot(x = counts2$DayNumber, y = counts2$Freq)
```

```r
counts3 <- dta3 %>% group_by_(.dots=colnames(dta3)) %>% summarise(Freq = n())
plot(x = counts3$DayNumber, y = counts3$Freq)
```

```r
counts4 <- dta4 %>% group_by_(.dots=colnames(dta4)) %>% summarise(Freq = n())
plot(x = counts4$DayNumber, y = counts4$Freq)
```

```r
counts5 <- dta5 %>% group_by_(.dots=colnames(dta5)) %>% summarise(Freq = n())
plot(x = counts5$DayNumber, y = counts5$Freq)
```

```r
counts6 <- dta6 %>% group_by_(.dots=colnames(dta6)) %>% summarise(Freq = n())
plot(x = counts6$DayNumber, y = counts6$Freq)
```

```r
counts7 <- dta7 %>% group_by_(.dots=colnames(dta7)) %>% summarise(Freq = n())
plot(x = counts7$DayNumber, y = counts7$Freq)
```

```r
write.csv(counts1, file = '../data/model_data1.csv', row.names = FALSE)
write.csv(counts2, file = '../data/model_data2.csv', row.names = FALSE)
write.csv(counts3, file = '../data/model_data3.csv', row.names = FALSE)
write.csv(counts4, file = '../data/model_data4.csv', row.names = FALSE)
write.csv(counts5, file = '../data/model_data5.csv', row.names = FALSE)
write.csv(counts6, file = '../data/model_data6.csv', row.names = FALSE)
write.csv(counts7, file = '../data/model_data7.csv', row.names = FALSE)
```
